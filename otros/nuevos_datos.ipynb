```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\rober\Desktop\Portafolio_Proyectos_RobertoGC97\credit_risk\data\datos_procesados.csv')
df.info()
```

```python
# Agregar cliente_id
df['cliente_id'] = range(1, len(df) + 1)

meses = ['abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre']

# Columnas fijas (no se derriten)
id_vars = ['cliente_id', 'linea_credito', 'genero', 'estado_civil', 
           'edad', 'DEFAULT', 'educacion_1', 'educacion_2']

# Melt de comportamiento
df_comp = df.melt(
    id_vars=id_vars,
    value_vars=[f'comportamiento_{mes}' for mes in meses],
    var_name='Mes', value_name='Comportamiento'
)
df_comp['Mes'] = df_comp['Mes'].str.replace('comportamiento_', '')

# Melt de estado_cuenta
df_estado = df.melt(
    id_vars=['cliente_id'],
    value_vars=[f'estado_cuenta_{mes}' for mes in meses],
    var_name='Mes', value_name='Estado_Cuenta'
)
df_estado['Mes'] = df_estado['Mes'].str.replace('estado_cuenta_', '')

# Melt de abono
df_abono = df.melt(
    id_vars=['cliente_id'],
    value_vars=[f'abono_{mes}' for mes in meses],
    var_name='Mes', value_name='Abono'
)
df_abono['Mes'] = df_abono['Mes'].str.replace('abono_', '')

# Unir todo
df_largo = df_comp.merge(df_estado, on=['cliente_id', 'Mes'])
df_largo = df_largo.merge(df_abono, on=['cliente_id', 'Mes'])

# Ordenar meses correctamente
orden_meses = {'abril': 1, 'mayo': 2, 'junio': 3, 'julio': 4, 'agosto': 5, 'septiembre': 6}
df_largo['Orden_Mes'] = df_largo['Mes'].map(orden_meses)
df_largo = df_largo.sort_values(['cliente_id', 'Orden_Mes'])

# Capitalizar meses
df_largo['Mes'] = df_largo['Mes'].str.capitalize()

# Exportar
df_largo.to_csv(r'C:\Users\rober\Desktop\Portafolio_Proyectos_RobertoGC97\credit_risk\data\powerbi_largo.csv', index=False)

print(df_largo.shape)
df_largo.head()
```
